# Usando Pandas para extrair dados por Web Scrape

```python
# Definindo os blocos de construção
ano = '2019'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html' 
# Unindo os blocos
url = url_link.format(ano)
url
```

```python
# Definindo os anos a ser analisados
anos = [2015,2016,2017,2018,2019]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
for ano in anos:
  url = url_link.format(ano)
  print(url)
```

```python
#Importando o Pandas
import pandas as pd
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
#Quantidade de tabela existente na pagina web:
len(df)
```

```python
df[0]
```

```python
df2019 = df[0]
```

```python
# Aplicando "Data cleaning"
df2019[df2019.Age == 'Age']
```

```python
len(df2019[df2019.Age == 'Age'])
```

```python
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
```
